## Streaming and Batch

### Streaming
 Most models can stream their output content while it is being generated.By displaying output progressively,steaming significantly improves user experience, particularly for longer responses. Calling stream() returns an iterator that yields output chunks as they are produced. You can use a loop to process each chunk in real-time.

```python
model.invoke("write me 200 words on Artificial intelligence")
# here we have to wait for complete answer
```

```python
for chunk in model.stream(" write me 200 words paragraph on Artificial Intellignce"):
    print(chunk.text, end="|",flush=True)
```

```python
for chunk in model.stream(" Why do parrots have colourful feathers?"):
    print(chunk.text, end="|",flush=True)
```

```python
model.invoke("Why do parrots have colourful feathers?")
```
